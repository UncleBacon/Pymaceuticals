```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse = "data/mouse_drug_data.csv"
clinical = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_df = pd.read_csv(mouse)
clinical_df = pd.read_csv(clinical)

# Combine the data into a single dataset
df = pd.merge(clinical_df, mouse_df, on = 'Mouse ID')
# Display the data table for preview
df.sort_values('Timepoint').reset_index(drop= True).head()
```

## Tumor Response to Treatment

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
# Convert to DataFrame
tumor = df.groupby(['Drug','Timepoint'])[['Tumor Volume (mm3)']].sem()
tumor = tumor.rename(columns=({'Tumor Volume (mm3)':'Tumor Volume SEM'}))


# Preview DataFrame
tumor2 = tumor.reset_index()
tumor2.head().style.format({'Tumor Volume SEM':'{:,.2f}'})
```

```python
# Minor Data Munging to Re-Format the Data Frames
agvol = df.groupby(['Timepoint','Drug'])['Tumor Volume (mm3)'].mean().unstack()

# Preview that Reformatting worked
agvol.head().style.format('{:,.2f}')
```

```python
# Generate the Plot (with Error Bars)
x_axis = agvol.index.values+5
cap = agvol['Capomulin']
inf = agvol['Infubinol']
ket = agvol['Ketapril']
pac = agvol['Placebo']

cap_graph, = plt.plot(x_axis,cap,marker = 'o',color='blue',label = "Capomulin")
inf_graph, = plt.plot(x_axis,inf,marker = '+',color='red',label = "Infubinol")
ket_graph, = plt.plot(x_axis,ket,marker = 'p',color='green',label = "Ketapril")
pac_graph, = plt.plot(x_axis,pac,marker = 'd',color='purple',label = "Placebo")
plt.xlabel('Timepoint\n(Days)') 
plt.ylabel("Tumor Volume (mm3)")
plt.title("Tumor Response to Treatment")
plt.grid(axis = 'y')
plt.legend(loc = 'best')
plt.savefig("images/Tumor_Response.png")
plt.show()
```

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mice = df.groupby(['Drug','Timepoint'])['Mouse ID'].count()
# Convert to DataFrame
mice_df = pd.DataFrame(mice)
mice_df = mice_df.rename(columns={'Mouse ID':'Mouse Count'}).reset_index()
# Preview DataFrame
mice_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mice2 = mice = df.groupby(['Timepoint','Drug'])['Mouse ID'].count().unstack()
# Preview the Data Frame
mice2.head()
```

```python
mice_df = mice2[['Capomulin','Infubinol','Ketapril','Placebo']]
mice_pct = mice_df/mice_df.max()


x_axis = mice_pct.index.values+5
cap = mice_pct['Capomulin']
inf = mice_pct['Infubinol']
ket = mice_pct['Ketapril']
pac = mice_pct['Placebo']

cap_graph, = plt.plot(x_axis,cap,marker = 'o',color='blue',label = "Capomulin")
inf_graph, = plt.plot(x_axis,inf,marker = '+',color='red',label = "Infubinol")
ket_graph, = plt.plot(x_axis,ket,marker = 'p',color='green',label = "Ketapril")
pac_graph, = plt.plot(x_axis,pac,marker = 'd',color='purple',label = "Placebo")
plt.xlabel('Timepoint\n(Days)') 
plt.ylabel("Survival Rate (%)")
plt.title("Survival Rate During Treatment")
plt.grid()
plt.legend(loc = 'best')
plt.savefig("images/Survival_Rate.png")
plt.show()
```
